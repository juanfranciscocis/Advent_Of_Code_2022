```python
stackA =['N','W','F','R','Z','S','M','D']
stackB = ['S','G','Q','P','W']
stackC = ['C','J','N','F','Q','V','R','W']
stackD = ['L','D','G','C','P','Z','F']
stackE = ['S','P','T']
stackF = ['L','R','W','F','D','H']
stackG = ['C','D','N','Z']
stackH = ['Q','J','S','V','F','R','N','W']
stackI = ['V','W','Z','G','S','M','R']

#invert the stacks
stackA = stackA[::-1]
stackB = stackB[::-1]
stackC = stackC[::-1]
stackD = stackD[::-1]
stackE = stackE[::-1]
stackF = stackF[::-1]
stackG = stackG[::-1]
stackH = stackH[::-1]
stackI = stackI[::-1]
```

```python
print(stackA)
print(stackB)
print(stackC)
print(stackD)
print(stackE)
print(stackF)
print(stackG)
print(stackH)
print(stackI)
```

```python
def chooseStack(number):
    if number == 1:
        return stackA
    elif number == 2:
        return stackB
    elif number == 3:
        return stackC
    elif number == 4:
        return stackD
    elif number == 5:
        return stackE
    elif number == 6:
        return stackF
    elif number == 7:
        return stackG
    elif number == 8:
        return stackH
    elif number == 9:
        return stackI
    else:
        return "Error"
```

```python
def moveFromStackToStack(instructions):
    for j in range(len(instructions)):
        quantity = int(instructions[j][0])
        stack1 = chooseStack(int(instructions[j][1]))
        stack2 = chooseStack(int(instructions[j][2]))
        if quantity == 1:
            stack2.append(stack1.pop())
        if quantity > 1:
            reverse = []
            for i in range(quantity):
                reverse.append(stack1.pop())
                stack2.append(reverse.pop())



        print(stackA)
        print(stackB)
        print(stackC)
        print(" ")



```

```python
moveFromStackToStack(instructions)
```

```python
#GET THE FISRT ITEM FROM THE STACKS IN A LIST
def getFirstItemFromStacks():
    firstItems = []
    firstItems.append(stackA.pop())
    firstItems.append(stackB.pop())
    firstItems.append(stackC.pop())
    firstItems.append(stackD.pop())
    firstItems.append(stackE.pop())
    firstItems.append(stackF.pop())
    firstItems.append(stackG.pop())
    firstItems.append(stackH.pop())
    firstItems.append(stackI.pop())
    return firstItems
```

```python
result = getFirstItemFromStacks()
```

```python
#convert the list to a string
result = ''.join(result)
print(result)
```
