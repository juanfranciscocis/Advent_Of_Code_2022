```python
with open('day4.txt', 'r') as file:
    data = file.read()

data = data.split("\n")
data
```

```python
for i in range(len(data)):
    data[i] = [x.split("-") for x in data[i]]
```

```python
# two lists with two numbers each
# check if the first list overlaps with the second list at all
def overlap(list1, list2):
    if list1[0] >= list2[0] and list1[0] <= list2[1]:
        return True
    if list1[1] >= list2[0] and list1[1] <= list2[1]:
        return True
    if list2[0] >= list1[0] and list2[0] <= list1[1]:
        return True
    if list2[1] >= list1[0] and list2[1] <= list1[1]:
        return True
    return False

```

```python
list1  = [6,6]
list2 = [4,6]
```

```python
overlap(list1, list2)
```

```python
list1  = [2,4]
list2 = [6,8]
```

```python
overlap(list1, list2)
```

```python
for i in range(len(data)):
    for j in range(len(data[i])):
        data[i][j] = [int(x) for x in data[i][j]]
```
